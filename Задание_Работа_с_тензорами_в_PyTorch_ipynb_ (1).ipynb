Задание 12. Создайте создайте булеву маску для массива "школьных оценок" из задания 10, в которой True будет стоять для первых 5 оценок и для последних 5 оценок, а остальные значения будут False.

```python
bool_mask = torch.zeros(len(C), dtype=torch.bool)  # Все False
bool_mask[:5] = True    # Первые 5 → True
bool_mask[-5:] = True   # Последние 5 → True
print(bool_mask)


```

Задание 13. Отфильтруйте тензор "школьных оценок", созданный в задании 10, с помощью маски, созданной в задании 12, оставив тем самым только первые и последние пять оценок.

```python
first_five = bool_mask[:5]
last_five = bool_mask[-5:]

# Объединяем в один массив
combined = torch.cat([first_five, last_five])

print("Объединенный массив (первые 5 + последние 5):")
print(combined)
```

Задание 14. Создайте маску для тензора "школьных оценок", в которой для каждой "пятерки" будет стоять True, а для остальных значений будет стоять False.

```python
P = torch.tensor(C)
for i in range(0,100):
  if P[i]==5:
    P[i]==1
  else:
    P[i]=0
P1 = torch.tensor(P,dtype=torch.bool)
print(P1)
```

Задание 17. Выполните стыковку тензоров a и b по вертикали.

Вход:

> a = torch.arange(2, 10, 2).reshape(2, -1)

> b = torch.repeat(1, 10).reshape(2, -1)

Желаемый результат: новый тензор размером $4 \times 5$.


```python
a = torch.arange(10).reshape(2, -1)
b = torch.ones(10).reshape(2, -1)
S = torch.cat([a, b], dim=0)#объединение по строкам (dim=0 вертикально)

print(
    f"Объединять вдоль существующей оси:\n{S}",
    f"Изменять размерность:\n{S.reshape(4, 5)}",
    sep="\n\n",

)
```

Задание 18. Выполните стыковку тензоров a и b по горизонтали.

Вход:

a = torch.arange(2, 10, 2).reshape(2, -1)

b = torch.repeat(1, 10).reshape(2, -1)

Желаемый результат: новый тензор размером 2 x 10

```python
a = torch.arange(2,10,2).reshape(2, -1)
b = torch.ones(1,10).reshape(2, -1)
V = torch.cat([a, b], dim=1)
print(V)
```
